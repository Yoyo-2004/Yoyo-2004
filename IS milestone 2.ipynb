PlayFair Password Encryption Implementation

```python
import string

def generate_playfair_matrix(key):
    key = key.lower().replace("j", "i")
    result = []
    used = set()

    for char in key + string.ascii_lowercase:
        if char not in used and char.isalpha():
            used.add(char)
            result.append(char)
        if len(result) == 25:
            break

    matrix = [result[i:i + 5] for i in range(0, 25, 5)]
    return matrix

def find_position(matrix, char):
    for i in range(5):
        for j in range(5):
            if matrix[i][j] == char:
                return i, j

def playfair_encrypt(text, key):
    text = text.lower().replace("j", "i")
    matrix = generate_playfair_matrix(key)

    prepared = ""
    i = 0
    while i < len(text):
        a = text[i]
        b = text[i + 1] if i + 1 < len(text) else "x"
        if a == b:
            prepared += a + "x"
            i += 1
        else:
            prepared += a + b
            i += 2

    encrypted = ""
    for i in range(0, len(prepared), 2):
        a, b = prepared[i], prepared[i+1]
        r1, c1 = find_position(matrix, a)
        r2, c2 = find_position(matrix, b)

        if r1 == r2:
            encrypted += matrix[r1][(c1+1)%5] + matrix[r2][(c2+1)%5]
        elif c1 == c2:
            encrypted += matrix[(r1+1)%5][c1] + matrix[(r2+1)%5][c2]
        else:
            encrypted += matrix[r1][c2] + matrix[r2][c1]

    return encrypted
```

VigenÃ¨re Cipher Encryption Implementation

```python
def vigenere_encrypt(text, key):
    text = text.lower()
    key = key.lower()
    encrypted = ""

    for i, char in enumerate(text):
        shift = ord(key[i % len(key)]) - ord('a')
        encrypted += chr((ord(char) - ord('a') + shift) % 26 + ord('a'))

    return encrypted
```

Bscrypt Encryption Implementation

```python
import bcrypt

def bcrypt_encrypt(password):
    password = password.encode()
    salt = bcrypt.gensalt()
    hashed = bcrypt.hashpw(password, salt)
    return hashed.decode()
```
